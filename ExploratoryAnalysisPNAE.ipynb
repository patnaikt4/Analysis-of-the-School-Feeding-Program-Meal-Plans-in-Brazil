```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Planilha2022_04_3_24.xlsx")
```

```python
df.head()
```

```python
df.columns = [
    "Year", "Executing_Entity", "State", "Transferred_Value",
    "Family_Agriculture_Value", "Percentage", "IBGE_Code", "Gov_Level", "Region"
]

df = df.drop(index=0).reset_index(drop=True)
```

```python
# Convert relevant columns to numeric (coerce errors to handle non-numeric values)
df["Transferred_Value"] = pd.to_numeric(df["Transferred_Value"], errors='coerce')
df["Family_Agriculture_Value"] = pd.to_numeric(df["Family_Agriculture_Value"], errors='coerce')
df["Percentage"] = pd.to_numeric(df["Percentage"], errors='coerce')
```

```python
summary = {
    "Total rows": len(df),
    "Rows with transfer value > 0": (df["Transferred_Value"] > 0).sum(),
    "Rows with family agriculture spending > 0": (df["Family_Agriculture_Value"] > 0).sum(),
    "Rows meeting 30% target": (df["Percentage"] >= 0.3).sum(),
    "Rows with 0% family farming spending": (df["Percentage"] == 0).sum()
}
```
